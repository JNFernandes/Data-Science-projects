##### There are 50 states of the US, Outside of the 50 main states in the Union, the United States has several territories in different parts of the world too. These territories may not be official states, but they still adhere to the same formula of two capital letters for their official abbreviations.

```python
state_list = pd.read_csv('List of States.csv')
```

```python
state_territories = pd.read_csv('US Commonwealth and Territories.csv')
```

```python
all_states = pd.concat((state_list,state_territories),axis=0,ignore_index=True)
```

```python
all_states 
```

```python
all_states['State']=all_states['State'].apply(lambda x: x.strip())
    
```

```python
def get_state(state_name):
    code = all_states[all_states['State']==state_name]['Code']
    return code.values[0].strip()
```

# Convert to SI units

```python
# 1 ft = 0.3048m 
# 1 sq. ft = 0.092903 m2
```

```python
bridge_data['Structure Length (ft)'] = round(bridge_data['Structure Length (ft)']*0.3048,0) 
bridge_data['Deck Area (sq.ft)'] = round(bridge_data['Deck Area (sq.ft)']*0.092903,0)
```

```python
bridge_data.head(5)
```

###### There are some names that could be changed in order to simplify the label name

```python
bridge_data.rename(columns={"Structure Length (ft)": "Bridge Length (m)", 
                            "Deck Area (sq.ft)": "Deck Area (m2)",
                           "Main Design Mat Type Value":"Material Type",
                           "Number of Spans in Main Unit":"No. of Spans",
                           }, errors="raise", inplace=True)
```

```python
bridge_data
```

#### Convert the no. of Spans, Bridge Age to integers

```python
bridge_data['No. of Spans'] = bridge_data['No. of Spans'].astype(int)
bridge_data['Bridge Age (yr)'] = bridge_data['Bridge Age (yr)'].astype(int)
```
